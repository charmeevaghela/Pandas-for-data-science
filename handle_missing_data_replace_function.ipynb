## here '-99999' represents 'no value'. so to show that in df use replase method

```python
new_df = df.replace(-99999,np.NaN)
new_df
```

## in case we have more 'no values' like '-9999' and '-8888'. 

```python
df = pd.read_csv("weather_data.csv")
df
```

```python
df1 = df.replace([-99999,-8888],np.NaN) #you can pass a list
df1
```

## in case when NaN values depends on the column. so there use dictionary

```python
df1 = df.replace(
    {
        "temperature": -99999,
        "windspeed":[-99999,-8888],
        "event": 'no event'
    },np.NaN
)
df1
```

## you can also pass simple dictionary. It is called mapping.

```python
df1 = df.replace({
    -99999:np.NaN,
    -8888:np.NaN,
    "no event":"not"
})
df1
```

## sometimes there are some unit of measures, like ^F, ^C, mph(miles per hour) etc.
## so you want to replace them with blank value. Use dictionary and regex for that

```python
df = pd.read_csv("weather_data.csv")
df
```

**by doing this units will be replaced with blank**

```python
df1 = df.replace({
    "temperature":'[A-Za-z]',
    "windspeed":'[A-Za-z]'
},"",regex=True)
df1
```

## Now replacing a list of values with another list of values.
## creating another df for that

```python
df = pd.DataFrame({
    'score':['exceptional','average','good','poor','average','exceptional'],
    'student':['rob','maya','parthiv','tom','julian','erica']
})
df
```

**provide 2 lists here, mapping values using list**

```python
df1 = df.replace(['exceptional','good','average','poor'],[1,2,3,4])
df1
```
